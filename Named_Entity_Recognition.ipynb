```python
# Import the dependencies
import spacy
from nltk.corpus import reuters
from spacy import displacy
from collections import Counter
import pandas as pd

# Load the small English language model for spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# Locate and store the article on public health.
with open('public_health.txt', 'r', encoding='utf-8') as file:
    text = file.read()

print(text)
```

```python
# Analyze the article with spacy
doc = nlp(text)

# Render NER visualization with displacy to determine entities for extraction
displacy.render(doc, style='ent')
```

Here's glossary of entities: 

| **Code**      | **Entity Type**                   | **Description**                                      |
| :------------ | :-------------------------------- | :--------------------------------------------------- |
| `PERSON`      | Person                            | People, including fictional characters               |
| `NORP`        | Nationality, Religious, Political | Nationalities, religious or political groups         |
| `FAC`         | Facility                          | Buildings, airports, highways, bridges, etc.         |
| `ORG`         | Organization                      | Companies, agencies, institutions                    |
| `GPE`         | Geopolitical Entity               | Countries, cities, states                            |
| `LOC`         | Location                          | Non-GPE locations (mountain ranges, bodies of water) |
| `PRODUCT`     | Product                           | Objects, vehicles, foods, etc. (not services)        |
| `EVENT`       | Event                             | Named hurricanes, battles, wars, sports events       |
| `WORK_OF_ART` | Work of Art                       | Titles of books, songs, artworks                     |
| `LAW`         | Law                               | Named documents like treaties, constitutions, laws   |
| `LANGUAGE`    | Language                          | Any named language                                   |
| `DATE`        | Date                              | Absolute or relative dates or periods                |
| `TIME`        | Time                              | Times smaller than a day                             |
| `PERCENT`     | Percentage                        | Percentage expressions                               |
| `MONEY`       | Money                             | Monetary values, including currency units            |
| `QUANTITY`    | Quantity                          | Measurements excluding time, money, percent          |
| `ORDINAL`     | Ordinal                           | “first”, “second”, etc.                              |
| `CARDINAL`    | Cardinal                          | Numerals not used as ordinals                        |
